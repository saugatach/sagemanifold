## Embed $\mathbb{S}^2 \rightarrow \mathbb{R}^3$

```python
reset()
S2 = Manifold(2, 'S2', r'\mathbb{S}^2')
polarS2.<th, ph> = S2.chart(r'th:(0,pi):\theta ph:(0,2*pi):\phi')
S2.atlas()
```

```python
polarS2.coord_range()
```

```python
R3 = Manifold(3, 'R3', r'\mathbb{R}^3')
cartR3.<x,y,z> = R3.chart()
cartR3.coord_range()
```

```python
embedding23 = S2.diff_map(R3,[sin(th)*cos(ph), sin(th)*sin(ph), cos(th) ], name="\Phi")
embedding23.disp()
```

The $S^2$ is plotted in $R^3$ using the differentiable map $\Phi$. The parameter **number_values** controls the density of the wire mesh.

```python
graph2 = polarS2.plot(cartR3, mapping=embedding23,  number_values=18)
```

```python
show(graph2, viewer='threejs')
```

## Embed $\mathbb{S}^4 \rightarrow \mathbb{R}^5$

```python
reset()
S4 = Manifold(4, 'S4', r'\mathbb{S}^4')
polarS4.<theta, phi, chi, psi> = S4.chart(r'theta:(0,pi) phi:(0,2*pi) chi:(0,2*pi) psi:(0,2*pi)')
S4.atlas()
```

Two possible maps can be used:
1. $z=sin(\theta)$ with $\theta \in (-\pi/2,\pi/2)$
2. $z=cos(\theta)$ with $\theta \in (\pi,0)$ 

The two maps are related by a coordinate change $\theta \rightarrow \pi/2 - \theta$

```python
polarS4.coord_range()
```

```python
R5 = Manifold(5, 'R5', r'\mathbb{R}^5')
cartR5.<z,x,y,u,v> = R5.chart()
cartR5.coord_range()
```

```python
embedding = S4.diff_map(R5,[cos(theta), sin(theta)*cos(phi), sin(theta)*sin(phi)*cos(chi), 
                              sin(theta)*sin(phi)*sin(chi)*cos(psi), sin(theta)*sin(phi)*sin(chi)*sin(psi) ])
embedding.disp()
```

We plot the $S^4$ in $R^5$. To plot in 3D we need to suppress the (u,v) coordinates (5D to 3D) and the corresponding polar coordinates $(\chi, \psi)$.

We do not need to explicitly suppress (u,v). Instead we declare the coordinate which will be used as coordinate axes: **ambient_coords=(x,y,z)**

We suppress the polar coordinates by **fixed_coords={chi:0, psi:0}**

```python
graph1 = polarS4.plot(cartR5, mapping=embedding, ambient_coords=(x,y,z), fixed_coords={chi:0, psi:0},
                      color={theta:'red', phi:'blue'}, number_values=18, max_range=2)
```

```python
show(graph1, viewer='threejs')
```

## Riemannian geometry 

#### Define a flat metric on $R^5$ and perform pullback of the metric onto $S^4$

```python
h = R5.metric('h')
for i in range(0,5):
    h[i,i]=1
h.disp()
```

```python
g=S4.metric('g')
```

```python
gh = embedding.pullback(h)
gh.disp()
```

The pullback function performs the following operations
1. Find differential of each coordinate of embedding space in terms of embedded coordinates as 1-forms
Example, $z = cos(\theta) \Rightarrow dz = -sin(\theta) d\theta$ 
2. Take inner product of the 1-forms to get bilinear forms
$ dz \otimes dz = sin^2(\theta) ~ d\theta \otimes d\theta  $
3. Combine the 2-forms in the metric to get the embedding metric
$ dx \otimes dx + dy \otimes dy + dz \otimes dz + du \otimes du + dv \otimes dv $

We need to set this pullback metric to the metric defined on $S^4$

```python
g.set(gh)
g.disp()
```

This is clearly the metric for $S^4$

## Embed $\mathbb{dS}^4 (\mathbb{R}^1 \times \mathbb{S}^3) \rightarrow \mathbb{R}^{1,4}$

```python
reset()
dS4 = Manifold(4, 'dS4', r'\mathbb{dS}^4')
polardS4.<tau, theta, phi, chi> = dS4.chart(r'tau theta:(0,2*pi) phi:(0,2*pi) chi:(0,2*pi)')
dS4.atlas()
```

Two possible maps can be used:
1. $z=sin(\theta)$ with $\theta \in (-\pi/2,\pi/2)$
2. $z=cos(\theta)$ with $\theta \in (\pi,0)$ 

The two maps are related by a coordinate change $\theta \rightarrow \pi/2 - \theta$

```python
polardS4.coord_range()
```

```python
R14 = Manifold(5, 'R14', r'\mathbb{R}^{1,4}')
cartR14.<t,x,y,u,v> = R14.chart()
cartR14.coord_range()
```

```python
embedding = dS4.diff_map(R14,[sinh(tau), cosh(tau)*cos(theta), cosh(tau)*sin(theta)*cos(phi), 
                              cosh(tau)*sin(theta)*sin(phi)*cos(chi),  cosh(tau)*sin(theta)*sin(phi)*sin(chi) ])
embedding.disp()
```

We plot the $sS^4$ in $R^5$. To plot in 3D we need to suppress the (u,v) coordinates (5D to 3D) and the corresponding polar coordinates $(\phi, \chi)$.

We do not need to explicitly suppress (u,v). Instead we declare the coordinate which will be used as coordinate axes: **ambient_coords=(x,y,t)** where t is along the z-axis.

We suppress the polar coordinates by **fixed_coords={chi:0, psi:0}**

```python
graph1 = polardS4.plot(cartR14, mapping=embedding, ambient_coords=(x,y,t), fixed_coords={phi:0, chi:0},
                      color={tau:'blue', theta:'blue'}, number_values=18, max_range=2)
```

```python
show(graph1, viewer='threejs')
```
